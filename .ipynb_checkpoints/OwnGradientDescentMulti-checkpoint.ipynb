```python
import numpy as np
from sklearn import datasets
def cost(x, y, m , b):
    return ((y - m * x.sum() - b) ** 2 ).mean()
```

```python
def step_gradient(x, y, rate, m, b):
    k = x.shape[0]
    slope_m = 0
    slope_b = 0
    for i in range(k):
        for j in range(x.shape[1]):
            slope_m += (2/k) * x[i][j] * (m * x[i][j] + b - y[i])
            slope_b += (2/k) * (m * x[i][j] + b - y[i])
        m = m - rate * slope_m
        b = b - rate * slope_b
    return m,b
```

```python
def gradient_descent(x, y, rate = 0.1, iterations = 100, start_m = 0, start_b = 0):
    m = np.array([0 for i in range(x.shape[0])])
    b = np.array([0 for i in range(x.shape[0])])
    #print("Start :", cost(x, y, m, b))
    for i in range(0, iterations):
        m,b = step_gradient(x, y, rate, m, b)
        #print(i, ":", cost(x, y, m, b))
    return m,b
```

```python
def run():
    boston = datasets.load_boston()
    x = boston.data[:, 0:3]
    y = boston.data[:, 12]
    rate = 0.000001
    iterations = 100
    m, b = gradient_descent(x, y, rate, iterations)
    print(m,b)
run()
```
