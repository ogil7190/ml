```python
import numpy as np
from sklearn import datasets
boston = datasets.load_boston()
```

```python
def predict(row, m):
    pred = m[0]
    for i in range(len(row)-1):
        pred += m[i + 1] * row[i]
    return pred
```

```python
def gradient_descent(x, y, rate, epoch):
    m = [0.0 for i in range(x.shape[1] + 1)]
    for i in range(epoch):
        for j in range(x.shape[0]):
            pred = predict(x[j], m)
            error = pred - y[j]
            m[0] = m[0] - rate * error
            for k in range(x.shape[1]):
                m[k+1] = m[k+1] - rate * error * x[j][k]
    return m
```

```python
def run():
    y = boston.target
    x = boston.data
    rate = 0.0000001
    epoch = 1000
    m = gradient_descent(x, y, rate, epoch)
    y_pred = list()
    for row in x:
        y_pred.append(predict(row, m))
    print(y_pred, y)
run()
```
