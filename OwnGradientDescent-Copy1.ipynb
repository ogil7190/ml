```python
import numpy as np
from sklearn import datasets
def cost(x, y, m, b):
    m = np.array(m)
    b = np.array(b)
    product = x.dot(m)
    value = ((y - product - b) ** 2 ).mean())
    return value
```

```python
def step_gradient(x, y, rate, m, b):
    p = x.shape[0]
    q = x.shape[1]
    new_m = m
    m = np.array(m)
    slope_m = 0
    slope_b = 0
    for i in range(q):
        for j in range(p):
            slope_m += (2/p) * x[j][i] * (m[i] * x[j][i] + b - y[i])
            slope_b += (2/p) * (m[i] * x[j][i] + b - y[i])
        new_m[i] = new_m[i] - rate * slope_m
        b = b - rate * slope_b
    return new_m,b
```

```python
def gradient_descent(x, y, rate = 0.1, iterations = 100, start_m = 0, start_b = 0):
    m = [0 for i in range(x.shape[1])]
    b = 0
    print("Start :", cost(x, y, m, b))
    for i in range(0, iterations):
        print("Before:", m, b)
        m,b = step_gradient(x, y, rate, m, b)
        print("After:", m, b)
        #print(i, ":", cost(x, y, m, b))
    return m,b
```

```python
def run():
    boston = datasets.load_boston()
    x = boston.data[:, 0:11]
    y = boston.data[:, 12]
    rate = 0.005
    iterations = 10
    m, b = gradient_descent(x, y, rate, iterations)
run()
```
