```python
import pandas as pd
import re
import math
```

```python
ndf = pd.read_csv('contacts2.csv')
for i in ndf:
    print(i)
```

```python
df = pd.read_csv('dad.csv')
phone = []

for i in range(len(df['Home Phone'])):
    n = df['Home Phone'][i]
    try:
        if(math.isnan(n)):
            phone.append('')
            continue
    except:
        n = n.replace('+91', '')
        n = re.sub("\D","", n)
        phone.append(n)

for i in range(len(df['Business Phone'])):
    m = df['Business Phone'][i]
    try:
        if(math.isnan(m)):
            if(len(phone[i]) > 0):
                continue
            else:
                phone[i] = ''
    except:
        m = m.replace('+91', '')
        m = re.sub("\D","", m)
        phone[i] = m

for i in range(len(df['Mobile Phone'])):
    n = df['Mobile Phone'][i]
    try:
        if(math.isnan(n)):
            if(len(phone[i]) > 0):
                continue
            else:
                phone[i] = ''
    except:
        n = n.replace('+91', '')
        n = re.sub("\D","", n)
        phone[i] = n
print(phone)
```

```python
col_names =  ['Name', 'Phone']
my_df  = pd.DataFrame(columns = col_names)
my_df['Name'] = df['Display Name']
my_df['Phone'] = phone
result = my_df.sort_values(['Name'])
result.to_csv('dad_output.csv', sep='\t', encoding='utf-8')
```
