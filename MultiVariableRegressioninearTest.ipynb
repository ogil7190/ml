```python
import numpy as np
from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
boston = datasets.load_boston()
```

```python
def cost(x, y, m):
    return np.sum((x.dot(m) - y) ** 2)/(2 * x.shape[0])
```

```python
def step_gradient(x, y, rate, m):
    p = x.shape[0]
    q = x.shape[1]
    slope_m = 0
    for i in range(q):
        for j in range(p):
            slope_m += (1/p) * x[j][i] * (m[i] * x[j][i] - y[i])
        m[i] = m[i] - rate * slope_m
    return m
```

```python
def gradient_descent(x, y, rate = 0.1, iterations = 100):
    m = np.zeros(x.shape[1])
    print("Start :", cost(x, y, m))
    for i in range(0, iterations):
        h = x.dot(m)
        loss = h - y
        gradient = x.T.dot(loss) / (2 * x.shape[0])
        m = m - rate * gradient
        #print(i, ":", cost(x, y, m))
    print(cost(x, y, m))
    return m
```

```python
def answer(m, x, y):
    ys = list()
    for i in range(x.shape[0]):
        y = 0
        for j in range(x.shape[1]):   
            y += x[i][j] * m[j]
        ys.append(y)
    return np.array(ys)
```

```python
def run(x, y):
    rate = 0.0000005
    iterations = 5000
    m = gradient_descent(x, y, rate, iterations)
    return m
```

```python
y = boston.target
x = boston.data
x = np.insert(x,0,[1],axis=1)
m = run(x, y)
print(m, boston.data.shape)
```

```python
pred1 = answer(m, x, y)
print(pred1)
```
